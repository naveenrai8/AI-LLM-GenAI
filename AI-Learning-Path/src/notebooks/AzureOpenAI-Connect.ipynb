```python
import os
from dotenv import load_dotenv
load_dotenv()
endpoint = os.getenv("AZURE_ENDPOINT")
model = os.getenv("AZURE_OPENAI_MODEL")

subscription_key = os.getenv("AZURE_OPENAI_API_KEY")
api_version = os.getenv("AZURE_OPENAI_API_VERSION")
```

<div class="alert alert-success"> 
1. Call to Azure OpenAI</br>
2. Record the time it takes.  </div>

```python
from openai import AzureOpenAI
import time

client = AzureOpenAI(
    api_version=api_version,
    azure_endpoint=endpoint,
    api_key=subscription_key,
)

start_time = time.perf_counter()

response = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Explain AI in 2 sentence.",
        }
    ],
    max_completion_tokens=16384,
    model=model
)
total_time = time.perf_counter() - start_time

print(response.choices[0].message.content)

```

```python
start_time = time.perf_counter()

response = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Explain AI in one sentence.",
        }
    ],
    model=model,
    stream=True,
    stream_options={
        "include_usage": True
    }
)

ttft = None
for chunk in response:
    if not ttft:
        ttft = time.perf_counter() - start_time
        print(f"[TTFT]: {ttft:.2f} seconds")
        
    if hasattr(chunk, 'usage') and chunk.usage:
        print(f"\n[Usage] Prompt Token: {chunk.usage.prompt_tokens}, Completion Token: {chunk.usage.completion_tokens}")

    elif chunk.choices and chunk.choices[0].delta.content:
        print(chunk.choices[0].delta.content, end="")
total_time = time.perf_counter() - start_time
```
