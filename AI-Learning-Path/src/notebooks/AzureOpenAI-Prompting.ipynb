```python

import os
from dotenv import load_dotenv
from openai import AzureOpenAI
import time

load_dotenv()
```

```python
class AzureOpenAIWrapper:
    def __init__(self, model):
        endpoint = os.getenv("AZURE_ENDPOINT")
        subscription_key = os.getenv("AZURE_OPENAI_API_KEY")
        api_version = os.getenv("AZURE_OPENAI_API_VERSION")
        self.model = model
        self.client = AzureOpenAI(
            api_version=api_version,
            azure_endpoint=endpoint,
            api_key=subscription_key,
        )

    def call_model_single_response(self, messages):
        start_time = time.perf_counter()

        response = self.client.chat.completions.create(
            messages=messages,
            max_completion_tokens=16384,
            model=self.model
        )
        total_time = time.perf_counter() - start_time
        print(response.choices[0].message.content)
```

```python
import os
model_wrapper = AzureOpenAIWrapper(os.getenv("AZURE_GPT_5_MINI_MODEL"))
model_wrapper.call_model_single_response([{"role":"user","content":"Explain CAP theorem in 3 sentences"}])

model_wrapper = AzureOpenAIWrapper(os.getenv("AZURE_OPENAI_MODEL"))
model_wrapper.call_model_single_response([{"role":"user","content":"Explain CAP theorem in 3 sentences"}])
```
