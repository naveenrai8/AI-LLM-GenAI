<h1>Tokenization</h1>

<div class="alert alert-success">Using OpenAI Tiktoken library</div>

```python
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-5-nano")
encoded = encoding.encode("Using tiktoken repo on Github for Tokenization to generate token.")
print(encoded)

decoded = encoding.decode(encoded)
print(decoded)
```

<div class="alert alert-warning">It generated different token id for same word with different case.</div>

<div class="alert alert-success">For input word:</br>
Github github gitHub git Hub</br>
[107290, 67154, 27339, 27052, 27339, 33512]</br>
Github -> 107290 </br>
github -> 67154</br>
gitHub -> 27339, 27052</br>

</div>
